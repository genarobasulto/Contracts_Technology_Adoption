## Introduction.


The adoption of new technology results from a series of individual decisions to begin using the new technology, decisions which are often the result of a comparison of the uncertain benefits of the new invention with the uncertain costs of adopting it. Importanlty, the choice is not only between adopting and not adopting but a choice between adopting now or deferring the decision until later when there is less uncertanty about the technology.

By and large, the benefits from adopting a new technology, are flow benefits which are received throughout the life of the acquired innovation. However, the costs are typically incurred at the time of adoption and cannot be recovered. That is, ex ante, a potential adopter weighs the fixed costs of adoption against the benefits he expects, but ex post, these fixed costs are irrelevant
because a great part of them have been sunk and cannot be recovered.

The benefits of adoption usually arise from network complementarities, for example which can affect the speed of difussion. For example, the benefits of owning a FCEV vehicle depend on the number of chargning stations in the network, the more chargning stations, the more attractive it is to own a FCEV. So, increasing the number of stations creates additional demand, which makes stations a better investment for suppliers. This network effect creates a feedback loop that accelerates the technology adoption. Network effect can be derived a more fundamental effect: mutual benefits of consumption. This allows consumers to care about the identity of other consumers and the amounts that they consume.

Although the ultimate decision is made on the demand side, the benefits and costs can be influenced by decisions made by suppliers of the new technology and the policy makers by, for example, giving a subsidy for the initial cost of the technology or, in the case of electric or hydrogen vehicles, subsidizing the construction of new stations.

Technology adoption usually follows a "S shape" across time.

## Model.
Suppose we have $\mathbf{M} = \{1,...,M\}$ different markets or sectors, each with $I_m$ distinct industrial consumers (if we want to work with continuum types, each market can have a mass $\phi_m$ of heterogeneous consumers with some distribution $f_m(.)$).

Time is discrete and finite $t = 1,...,T$. At each time $t$, a firm that produces an alternative fuel (hydrogen) makes two decisions: 1) whether to enter a new market at a cost $C_{mt}$ and 2) which contracts $\{(q_{imr}, p_{imr})\}_{r \geq t}$ to offer different consumers. Think about entering a new market as setting a new hydrogen station near a certain city (or port), the cost $C_{mt}$ captures the investment of bulding this station, which could partially be covered by a government subsidy. The contracts specify fuel quantities $q_{imr}$ that the companny promises to supply customer $i$ in market $m$ at a price $p_{imr}$, from the moment they contract onwards. Let $N_t$ the number of markets that the firm serves at time $t$.

The consumers then decide wheter to switch to the alternative fuel at an initial cost $P_t$ and conctract with the supplier, or continue using their current technology and receive certain utility $\bar{u}_{imt}$. Consumer utility (profit) from switching to the alternative fuel at time $t$ and contracting with the supplier is given by
$$
u_{imt} = V_{imt}(q_{imt}, N_t) - P_t  -  \sum_{r = t }^T q_{imr} p_{imr}  +   \sum_{r = t + 1}^T E_{N_r} [V_{im}(q_{imr}, N_r)]
$$
Where $V_{imt}$ is the flow benefit of owning the alternative fuel vehicle when the network size is $N$.

Let $1_{mt} = 1$ if the supplier firm enters market $m$ at time $t$ and 0 otherwise. And $A_{mt} = 1$ if the firm is active in market $m$ at time $t$. We can write down the firm's problem at time $t = 1$ as: (NOTE: This notation might not be the best/might not capture the full problem)
\begin{align*}
\max_{\{1_{mt}, (q_{imt}, p_{imt})\}_{t =1,...,T}} & \left\lbrace \underbrace{ \sum_{t = 1}^T\sum_{m = 1}^M \sum_{i = 1}^{I_m} A_{mt} q_{imt}(p_{imt} - c_{t}) }_{\text{Flow revenue}} - \underbrace{\sum_{t = 1}^T\sum_{m = 1}^M 1_{mt}C_{mt}}_{\text{Initial fixed cost}} \right\rbrace \\
\text{subject to} & \hspace{0.5cm} u_{imt} = V_{imt}(q_{imt}, N_t) - P_t  -  \sum_{r = t }^T q_{imr} p_{imr}  +   \sum_{r = t + 1}^T E_{N_r} [V_{im}(q_{imr}, N_r)] \geq \bar{u}_{imt}, \hspace{1cm} \forall i,m,t \\
& \hspace{0.5cm} N_t = \sum_{m = 1}^M A_{mt}
\end{align*}

We can write this problem recursively as
\begin{align*}
\Pi^*_t(A_{mt-1}, \{(q_{imr}, p_{imr})\}_{r =t-1,...,T}) =  \max_{1_{mt}, \{(q_{imr}, p_{imr})\}_{r =t,...,T}} & \left\lbrace \underbrace{\sum_{m = 1}^M \sum_{i = 1}^{I_m} A_{mt} q_{imt}(p_{imt} - c_{t}) }_{\text{Flow revenue}} - \underbrace{\sum_{m = 1}^M 1_{mt}C_{mt}}_{\text{Initial fixed cost}} \right\rbrace + \Pi^*_{t+1}(A_{mt},  \{(q_{imr}, p_{imr})\}_{r =t,...,T}) \\
\text{subject to} & \hspace{0.5cm} u_{imt} = V_{imt}(q_{imt}, N_t) - P_t  -  \sum_{r = t }^T q_{imr} p_{imr}  +   \sum_{r = t + 1}^T E_{N_r} [V_{im}(q_{imr}, N_r)] \geq \bar{u}_{imt}, \hspace{1cm} \forall i,m \\
& \hspace{0.5cm} N_t = \sum_{m = 1}^M A_{mt-1} + \sum_{m=1}^M 1_{mt}
\end{align*}

In the former recursive formulation, the relevant state variables are the set of markets in which the firm was active at time $t-1$ and the contracts it has to supply to existing customers. It decides the new markets to enter and new constracts to sign.

**Assumptions on the parameters.**

1) The cost of entering a market, $C_{mt}$ is weakly decreasing in time (or maybe decreasing in the size of the network in a future version),

2) The initial cost of the new technology $P_t$ for the industrial customers is weakly decreasing in time,

3)  What happens to expectations? I'm thinking they might not be rational but bounded by $[N_t, M]$ at each time.

Notice that in this way the supplier firm has some incentives to delay entry in some markets as it will face lower costs. Hoever, this will imply a loss of revenue for some periods. The customers also have an incentive to adopt the technology later rather than sooner as they might face lower costs and uncertanty.  

## Solution.

**Proposition 1.** Suppose consumers have rational expectation in equilibrium, i.e.
$$E_{N_r} [V_{im}(q^*_{imr}, N^*_r)] = V_{im}(q^*_{imr}, N^*_r)$$
And suppose the outside option $\bar{u}_{imt} = 0$. If the firm enter a market $m$ at time $t$, it will offer contract to consumer $i$ whenever
\begin{align*}
V_{im}(q^*_{imr}, N^*_r) - P_t + P_{t+1} \geq 0
\end{align*}
The optimal contract $\{(q^*_{imr}, N^*_r)\}_{r\geq t}$ satisfies
$$V'(q_{imr}^*, N_r^*) - c_t = 0 $$
And
$$p^*_{imt} = \frac{V_{im}(q^*_{imt}, N^*_t)-P_t}{q_{imt}^*},$$
$$p^*_{imr} = \frac{V_{im}(q^*_{imr}, N^*_r)}{q_{imr}^*}, \hspace{0.5cm} r > t$$
The buyer accepts the contract.

**Proof.** First notice that it is optimal for the firm to set the buyer's participation contraint to 0. If the contract is offereded at timne t, this implies
$$ V_{imt}(q_{imt}^*, N_t^*) - P_t  -  \sum_{r = t }^T q^*_{imr} p_{imr}^*  +   \sum_{r = t + 1}^T V_{im}(q_{imr}^*, N_r^*) = 0$$
$$⇒ \sum_{r = t }^T q^*_{imr} p_{imr}^* =  \sum_{r = t}^T V_{im}(q_{imr}^*, N_r^*)  - P_t$$
If the firms waits to period $t+1$ to offer the contract
 $$\sum_{r = t+1}^T q^*_{imr} p_{imr}^* =  \sum_{r = t+1}^T V_{im}(q_{imr}^*, N_r^*)  - P_{t+1}$$
The revenue from the contract starting at time $t$ minus that of the contract starting at time $t+1$ is
$$q_{imt}^*p_{imt}^* = V_{im}(q_{imt}^*, N_r^*) - P_{t} + P_{t+1}$$
So, offering the contract at time t is more profitable if $V_{im}(q_{imt}^*, N_r^*) - P_{t} + P_{t+1}>0$.

The second part of the proposition comes from the FOC of the firm after plugging in the revenue from the participation constraint. The buyers accept the contract by construction given it gives them at least their reservation utility.

The result can be interpreted intuitively: optimal contracts equate the flow marginal revenue with marginal costs. When deciding whether to offer a contract at time $t$ or defer it by one more period, the firm forgoes today's flow revenue, $V(q^*_t, N^*_t)$, but can extract additional revenue tomorrow, $P_t - P_{t+1} > 0$. As long as the loss of today's flow revenue exceeds the potential gains from lower technology costs in the future, the firm prefers to offer contracts early.

Another observation we can make from proposition 1 is that even when the firm makes the initial investment in one market, it might not want to offer all consumers in that market a contract and would rather defer the adoption until the switch to the new technology is cheaper for some customers or the network effects make the switching more attractive (for consumers with lower valuation for the fuel).

Solving the firm's entry problem is complex, using the recusive formulation of the problem, if at any time $t$ the firm can choose $K$ markets to enter then we will have to check $2^K$ possible combinations of markets (e.g. enter all markets, enter only market 1, enter market 1 and 3, enter market 3,4,5, etc.). Then, we have to carry back all the $2^K$ optimization problems backwards in time in order to give the continuation value at time $t-1$. One extra assumption that could solve this issue is to have the firm only being able to enter 1 market at the time, in that way we only need to solve for $K$ cases. ## An example.

Suppose there are $t =1,2, 3$ periods of time and $\mathbb{M} = \{1,2\}$ markets, each with 2 consumers with utility
$$V_{im}(q, N) = Nq (\theta_i - q)$$
with $\theta_1 = 2, \theta_2 = 1$ for consumer 1 and 2 in each market. This means each market has one consumer with high valuation for the technology and one with low valuation. The result from proposition 1 implies that the optimal quantity to offer each consumer conditional on the firm entering the market is
\begin{align*}
q_{imt}^* = \theta_i- c, \text{ if }  Nc(\theta_i-c) + N- P_{t} + P_{t+1} > 0
\end{align*}
In this example the firm can choose to enter both markets at the same time.

For simplicity, assume marginal costs are constant $c_t = 1$. Entering market 2 is more expensive at all times than entering market 1 and the initial cost of adopting the new technology for the consumers is $P_T = [2.3, 1,1]$.

The following program solves the firm's contracting problem:




```python
import numpy as np


#Parameters
I1 = np.array([2, 1.5]) #Consumer types mkt 1
I2 = np.array([2, 1.5]) #Consumer types mkt 2
Pt = np.array([1, 0.8, 0.3]) #Technology entry cost
DPt = Pt - np.array([1, 1, 1]) #Diff of technology entry cost.
ct = np.array([1, 1 , 1]) #Marginal cost
C1 = np.array([1,0.7, 0.5]) #Entry cost mkt 1
C2 = np.array([3, 2, 1]) #Entry cost mkt 2

#Useful functions
def v(q,N, theta):
  return N*q*(theta - q)
def ret_P(OC, Pt):
  P = np.zeros((3,2))
  for k in range(0,3):
    for i in range(0,2):
      if OC[k, i] == True:
       P[k,i] = Pt[i]
       break
  return P

q1 = np.array([I1-c for c in ct]) #Optimal quantities for market 1 at time t
q2 = np.array([I2-c for c in ct])

R1_1 = v(q1, 1, I1) #Revenues at mkt 1 when its the only market operating
R1_2 = v(q1, 2, I1) #Revenues at mkt 1 when both markets are operating
#Indicator of wheter to offer contract
OC1_1 = np.transpose(np.array([R1_1[:,i]  > DPt for i in range(0,2)]))
OC1_2= np.transpose(np.array([R1_2[:,i]  > DPt for i in range(0,2)]))

R2_1 = v(q2, 1, I2) #same but for mkt 2
R2_2 = v(q2, 2, I2)
OC2_1 = np.transpose(np.array([R2_1[:,i]  > DPt for i in range(0,2)]))
OC2_2= np.transpose(np.array([R2_2[:,i]  > DPt for i in range(0,2)]))
```

```python
#Tecnhology prices
P1_1 = ret_P(OC1_1, Pt)
P1_2 = ret_P(OC1_2, Pt)
P2_1 = ret_P(OC2_1, Pt)
P2_2 = ret_P(OC2_2, Pt)
```

```python
Pi = np.zeros((3,3)) #This will save optimal profit Pi_{tr} when firm enters mkt 1 at t and 2 at r
for t1 in range(0,3):
  for t2 in range(0,3):
    Rev1_1 = (R1_1 - P1_1)*OC1_1
    Rev1_2 = (R1_2 - P1_2)*OC1_2
    Rev2_1 = (R2_1 - P2_1)*OC2_1
    Rev2_2 = (R2_2 - P2_2)*OC2_2

    if t1<=t2: #If market 1 is the first to enter
      Pi[t1,t2] = np.sum(Rev1_1[t1:t2]) + np.sum(Rev1_2[t2:]) + np.sum(Rev2_2[t2:]) - C1[t1] - C2[t2] #Total profit
    else: #If market 2 is the first to enter
      Pi[t1,t2] = np.sum(Rev1_2[t1:]) + np.sum(Rev2_2[t1:]) + np.sum(Rev2_1[t2:t1]) - C1[t1] - C2[t2]
```

```python
Pi #The maximum happens when the firm enters market 1 at time 1 and market 2 at time 2.
```

```python
t1_opt = 0
t2_opt = 1
Pi[t1_opt, t2_opt]
```

For this specific set of parameters, both markets are equal in terms of their demand, so the firm first enters the market with the lowest entry costs (Market 1), it offers a contract to the customer with **lowest** valuation for the good (notice demand is negatively correlated with the valuation), the firm cannot extract much from the high valuation customers given they demand less and the cost of the technology is high. In period 2, the firm enters market 2 as the entry cost has decreased. The network effects and the lower costs of technology adoption allow the firm to capture the reminder of market 1 as well as the total demand of market 2.

```python
def extract_solution(t1, t2):
  if t1<=t2: #If market 1 is the first to enter
      Rev1  = np.concatenate([Rev1_1[t1:t2], Rev1_2[t2:]])
      Rev2  = np.concatenate([np.zeros((3,2))[t1:t2], Rev2_2[t2:]])
      Q1  = np.concatenate([(q1*OC1_1)[t1:t2], (q1*OC1_2)[t2:]])
      Q2  = np.concatenate([np.zeros((3,2))[t1:t2], (q2*OC2_2)[t2:]])
  else: #If market 2 is the first to enter
    Rev2  = np.concatenate([Rev2_1[t2:t1], Rev2_2[t1:]])
    Rev1  = np.concatenate([np.zeros((3,2))[t2:t1], Rev1_2[t1:]])
    Q2  = np.concatenate([(q2*OC2_1)[t2:t1], (q2*OC2_2)[t1:]])
    Q1  = np.concatenate([np.zeros((3,2))[t2:t1], (q1*OC1_2)[t1:]])
  return Rev1, Rev2, Q1, Q2
```

```python
R1_opt, R2_opt, q1_opt, q2_opt = extract_solution(t1_opt, t2_opt)
```

```python
Q = np.sum(q1_opt) + np.sum(q2_opt) #Total quantity solc
Fq1 = np.cumsum(np.sum(q1_opt, axis = 1))/Q #Cummulative total fraction of Q
Fq1 = np.insert(Fq1, 0,0)
Fq2 = np.cumsum(np.sum(q2_opt, axis = 1))/Q
Fq2 = np.insert(Fq2, 0,0)
time = np.array([0, 1,2,3])
I = np.array([0, 1, 4, 4])
print("Total quantity sold: ", Q)
```
